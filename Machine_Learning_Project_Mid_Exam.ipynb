###**3.1 Apakah ada hubungan antara response dan predictors? (Tes null hypothesis, hitung F-statistic atau p-value)**

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from scipy import stats

laptop = laptops
X = laptop[["Storage","Manufacturer","Screen Size"]]  # Nama predictor
y = laptop["Price"]  # Nama response

# Membagi dataset nya menjadi 2 jenis data: Data training & Data testing
x_training, x_testing, y_training, y_testing = train_test_split(X, y, test_size=0.4, random_state=42)

# Membuat Regression KNN Model, dengan jumlah k/neighbor = 5
model_knn = KNeighborsRegressor(n_neighbors=5)  #  n_neighbors = k = jumlah neighbor/tetangga dari KNN Model

# Melakukan Training/pelatihan model pada dataset
model_knn.fit(x_training, y_training)

y_predict = model_knn.predict(x_testing)

# Menghitung nilai F-statistic dan p-value:
f_statistic, p_value = stats.f_oneway(y_testing, y_predict)

# Print hasil testing dataset
print("F-statistic:", f_statistic)
print("P-value:", p_value)

# Menentukan tingkat signifikansi/significance value (alpha)
alpha = 0.05 # 5%

# Membandingkan nilai p-value dengan significance value(alpha)
if p_value < alpha:
    print("Terdapat hubungan yang significance antara variabel response dan predictors.")
else:
    print("Tidak terdapat hubungan yang significance antara variabel response dan predictors.")
```

### **3.2 Apakah semua predictors penting untuk memprediksi response, atau hanya sebagian saja? (Cek Cp, AIC, BIC, atau adjusted R2 pada test data untuk subset data tertentu)**

```python
import numpy as np
import statsmodels.api as sm
from itertools import combinations

# Data
X = laptop[["Storage","Manufacturer","RAM","Weight","Screen Size"]]
y = laptop["Price"]

# Inisialisasi variabel
best_aic = np.inf
best_model = None
predictor_names = X.columns

# Lakukan model selection menggunakan AIC
for k in range(1, len(predictor_names) + 1):
    for combo in combinations(predictor_names, k):
        combo = list(combo)
        X_sm = sm.add_constant(X[combo])
        model = sm.OLS(y, X_sm).fit()
        current_aic = model.aic
        if current_aic < best_aic:
            best_aic = current_aic
            best_model = model
            best_predictors = combo

# Print model terbaik dan fitur yang dipilih
print("Best Model AIC:", best_aic)
print("Best Predictors:", best_predictors)
print(best_model.summary())
```

# **4. Resampling Methods**

```python
import pandas as pd
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split, KFold

laptop = laptops
x = laptop[["Storage","Manufacturer","Screen Size"]]  # Variabel predictor
y = laptop["Price"]  # Variabel response

# Melakukan split data pada data training dan data testing:
x_training, x_testing, y_training, y_testing = train_test_split(x, y, test_size=0.25)

# Melakukan pengujian/fitting pada Model KNN Regression:
model_validation = KNeighborsRegressor(n_neighbors=5).fit(x_training, y_training)

# Mengkalkulasikan nilai Mean Squared Error/MSE pada dataset:
mse_validation = model_validation.score(x_testing, y_testing)

# Melakukan cross validation dengan metode K-Fold Cross validation:
kFold = KFold(n_splits=5, shuffle=True)

# Mengkalkulasikan nilai MSE/Mean Squared Error pada tiap Fold:
mse_kfold = []
for train_index, test_index in kFold.split(x):
    x_train_kfold, x_test_kfold = x.iloc[train_index], x.iloc[test_index]
    y_train_kfold, y_test_kfold = y.iloc[train_index], y.iloc[test_index]

    # Melakukan Fitting pada Model KNN Regression:
    model_kfold = KNeighborsRegressor(n_neighbors=5).fit(x_train_kfold, y_train_kfold)

    # Melakukan kalkulasi nilai MSE pada dataset:
    mse_kfold.append(model_kfold.score(x_test_kfold, y_test_kfold))

# Menampilkan nilai MSE untuk metode validation test dan K-Fold Cross Validation:
print("MSE dengan metode Validation Set:", mse_validation)
print("MSE dengan metode K-Fold Cross Validation:", np.mean(mse_kfold))

# Membahas nilai bias-variance trade off
# Metode Validation set:
# Bias Trade off: nilai bias trade off nya rendah, sehingga model dapat menampung pola data dengan baik
# Variance: nilai variance nya tinggi, sehingga model sensitif terhadap fluktuasi data
# Metode K-fold cross validation:
# Bias Trade off: nilai bias trade off nya rendah, sehingga model dapat menampung pola data dengan baik
# Variance: nilai variance nya rendah, sehingga model tidak sensitif terhadap fluktuasi data
```
